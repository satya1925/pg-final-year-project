```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
X = np.linspace(0, 10, 50)
y = 2.5 * X + np.random.randn(50) * 2

# Initialize parameters
m, c = 0.0, 0.0
lr = 0.01
epochs = 1000

# Gradient Descent
for i in range(epochs):
    y_pred = m * X + c
    dm = (-2/len(X)) * sum(X * (y - y_pred))
    dc = (-2/len(X)) * sum(y - y_pred)
    m -= lr * dm
    c -= lr * dc

print(f"Learned Parameters: m = {m:.3f}, c = {c:.3f}")

# Plot Results
plt.scatter(X, y, color='blue')
plt.plot(X, m*X + c, color='red')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np

# Sigmoid Activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Input and Output data
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Initialize weights and bias
np.random.seed(42)
weights = np.random.uniform(size=(2,1))
bias = np.random.uniform(size=(1,))
lr = 0.1

# Training Loop
for epoch in range(10000):
    # Forward Pass
    z = np.dot(X, weights) + bias
    output = sigmoid(z)

    # Backward Pass
    error = y - output
    d_output = error * sigmoid_derivative(output)

    weights += lr * np.dot(X.T, d_output)
    bias += lr * np.sum(d_output)

print("Final Weights:\n", weights)
print("Final Bias:", bias)
print("Predicted Output:\n", output.round(3))
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import numpy as np

data = load_iris()
X, y = data.data, data.target

model = LogisticRegression(max_iter=200)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(model, X, y, cv=kfold)

print("Cross Validation Scores:", scores)
print("Mean Accuracy:", np.mean(scores))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Generate dataset
X, y = make_moons(n_samples=1000, noise=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Normalize
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Model
model = Sequential([
    Dense(16, activation='relu', input_dim=2),
    Dropout(0.3),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), verbose=0)

loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Model Accuracy: {acc*100:.2f}%")
```

```python
from collections import defaultdict

def mapper(text):
    words = text.strip().split()
    return [(word.lower(), 1) for word in words]

def reducer(mapped_data):
    word_count = defaultdict(int)
    for word, count in mapped_data:
        word_count[word] += count
    return word_count

if __name__ == "__main__":
    text = "Data mining is about discovering patterns in large data sets"
    mapped = mapper(text)
    reduced = reducer(mapped)
    print("Word Count:", dict(reduced))
```

```python
from itertools import combinations

def apriori(transactions, min_support):
    itemsets = {}

    for transaction in transactions:
        for item in transaction:
            itemsets[frozenset([item])] = itemsets.get(frozenset([item]), 0) + 1

    num_transactions = len(transactions)
    current_itemsets = {k: v for k, v in itemsets.items() if v / num_transactions >= min_support}
    all_frequent = dict(current_itemsets)
    k = 2

    while current_itemsets:
        candidates = [i.union(j) for i in current_itemsets for j in current_itemsets if len(i.union(j)) == k]
        candidates = list(set(candidates))
        item_counts = {c: 0 for c in candidates}

        for transaction in transactions:
            t = set(transaction)
            for c in candidates:
                if c.issubset(t):
                    item_counts[c] += 1

        current_itemsets = {k: v for k, v in item_counts.items() if v / num_transactions >= min_support}
        all_frequent.update(current_itemsets)
        k += 1

    return all_frequent

transactions = [
    ['milk', 'bread', 'butter'],
    ['bread', 'butter'],
    ['milk', 'bread'],
    ['milk', 'bread', 'butter', 'jam']
]

frequent = apriori(transactions, 0.5)
print("Frequent Itemsets:", frequent)
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = np.array([
    [1, 2], [1.5, 1.8], [5, 8],
    [8, 8], [1, 0.6], [9, 11]
])

kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

print("Cluster Centers:\n", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='X')
plt.title("K-Means Clustering")
plt.show()
```

```python
import hashlib
import numpy as np

class CountMinSketch:
    def __init__(self, width, depth):
        self.width = width
        self.depth = depth
        self.table = np.zeros((depth, width), dtype=int)

    def _hash(self, item, i):
        return int(hashlib.md5((str(item) + str(i)).encode()).hexdigest(), 16) % self.width

    def add(self, item):
        for i in range(self.depth):
            self.table[i][self._hash(item, i)] += 1

    def count(self, item):
        return min(self.table[i][self._hash(item, i)] for i in range(self.depth))

cms = CountMinSketch(width=10, depth=3)
stream = ['apple', 'banana', 'apple', 'orange', 'apple', 'banana']

for item in stream:
    cms.add(item)

print("Approx Count of 'apple':", cms.count('apple'))
print("Approx Count of 'banana':", cms.count('banana'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

data = {
    'Movie': ['Inception', 'Interstellar', 'Dark Knight', 'Avatar', 'Titanic'],
    'Tags': [
        'sci-fi dream mind thriller',
        'space sci-fi future mission',
        'hero action gotham crime',
        'fantasy alien planet',
        'romance ship ocean love'
    ]
}

df = pd.DataFrame(data)
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['Tags'])
similarity = cosine_similarity(count_matrix)

def recommend(movie):
    idx = df[df.Movie == movie].index[0]
    scores = list(enumerate(similarity[idx]))
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:3]
    print(f"Movies similar to '{movie}':")
    for i, _ in sorted_scores:
        print(" ", df.Movie[i])

recommend('Inception')
```
